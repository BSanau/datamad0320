#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)

#plt.figure(figsize = (10,8))
plt.subplot(2,1,1) # equivalente a plt.subplot(211) - filas, columnas e índice de la gráfica
plt.plot(x,y, color='cyan', marker='o', linestyle='dashed', linewidth=1, markersize=2)
plt.subplot(2,1,2)
plt.plot(x,z, color='#008B8B', marker='s', linestyle='solid', linewidth=2, markersize=3)
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)

figure, (s1, s2) = plt.subplots(2,1) # colocadas en 2 filas de 1 columna
s1.plot(x,y, color='m', marker='+', linestyle='dotted', linewidth=1, markersize=2)
s2.plot(x,z, color="#F08080", marker='^', linestyle='None', linewidth=1, markersize=3)
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here
figure, sbplt = plt.subplots(2 , figsize = (10,8))
sbplt[0].plot(x,y, "b--1")
sbplt[1].plot(x,z, "y_")
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here

figure, (s1, s2) = plt.subplots(nrows = 2, ncols = 1, figsize = (10,7))
s1.plot(x,y, color='m', marker='+', linestyle='dotted', linewidth=1, markersize=2)
s1.set_xlabel("aa", fontsize=14)
s1.set_ylabel("bb", fontsize=14)
s2.plot(x,z, color="#F08080", marker='^', linestyle='None', linewidth=1, markersize=3)
s2.set_xlabel("cc", fontsize=14)
s2.set_ylabel("dd", fontsize=14)
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
plt.plot(x, x**2)
plt.plot(x, np.exp(x))
plt.yscale("log") # Por defecto, la escala es normal y aquí especificamos que la y sea logarítmica
```
