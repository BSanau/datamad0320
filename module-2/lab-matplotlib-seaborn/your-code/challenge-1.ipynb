#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here

figure, (s1, s2) = plt.subplots(nrows = 2, ncols = 1, figsize = (10,7))

s1.plot(x,y, color='m', marker='+', linestyle='dotted', linewidth=1, markersize=2)
s1.set_xlabel("x", fontsize=14)
s1.set_ylabel("y", fontsize=14)

s2.plot(x,z, color="#F08080", marker='^', linestyle='None', linewidth=1, markersize=3)
s2.set_xlabel("x", fontsize=14)
s2.set_ylabel("z", fontsize=14);
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here

figure, subplt = plt.subplots(1,2, figsize=(12,6))

# Con escala normal
subplt[0].plot(x, x**2)
subplt[0].plot(x, np.exp(x))
subplt[0].set_xlabel("x")
subplt[0].set_ylabel("y")
subplt[0].set_title("Normal scale")

# Con escala logarítmica
subplt[1].plot(x, x**2)
subplt[1].plot(x, np.exp(x))
subplt[1].set_yscale("log") # Por defecto, la escala es normal y aquí especificamos que la y sea logarítmica
subplt[1].set_xlabel("x")
subplt[1].set_ylabel("y")
subplt[1].set_title("Log scale");
```

#### In the vehicles data set that you have downloaded, use the vehicles.csv file. In this exercise we will conduct some exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. 

**Scatter Plot**

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
# your code here

vehicles = pd.read_csv("vehicles.csv")
vehicles.head()
```

```python
# Scatter plot con plt

plt.scatter(vehicles["Combined MPG"], vehicles["Highway MPG"], c="y", marker = "p", edgecolors ="g")
plt.xlabel("Combined MPG")
plt.ylabel("Highway MPG")
plt.title("Vehicles");
```

```python
# Scatterplot con sns

sns.scatterplot(x="Combined MPG", y="Highway MPG", data=vehicles);
```

**Box Whisker Plot**

Please provide a box plot of the variable "CO2 Emission Grams/mile"

```python
# Boxplot con plt

plt.boxplot(vehicles["CO2 Emission Grams/Mile"], vert=True, labels=["CO2 g/mi"])
plt.title("Boxplot CO2");
```

```python
# Boxplot con sns

sns.boxplot(vehicles["CO2 Emission Grams/Mile"], linewidth=1, orient="h", palette="Set2").set_title("Boxplot CO2");
```

**Histogram**

Please provide a histogram of the Fuel Barrels/Year

```python
# Histogram con plt

plt.style.use('ggplot') #"style" le da estilo al histograma
plt.hist(vehicles["Fuel Barrels/Year"], bins = 15)
plt.xlabel("Fuel Barrels/Year")
plt.ylabel("Frequency")
plt.title("Fuel histogram");
```

```python
# Histograma con sns
# Dibuja el histograma y la curva de densidad

sns.distplot(vehicles["Fuel Barrels/Year"]).set_title("Fuel histogram");
```

**Bar Chart**

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
x = vehicles["City MPG"].groupby(vehicles["Fuel Type"]).mean()
x.plot(kind = "bar")
plt.xlabel("Fuel Type")
plt.ylabel("City MPG")
plt.title("Bar chart");
```
