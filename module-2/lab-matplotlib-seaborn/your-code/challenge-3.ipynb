#### Use the appropriate visulalization to show the distribution of the column `Age`

```python
# Plotting histogram with plt

plt.hist(titanic["Age"], bins = 20)
plt.xlabel("Ages")
plt.ylabel("Frequency")
plt.title("Age histogram");
```

#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
# your code here
figure, sub = plt.subplots(3)
Nbins = [10, 20, 50]
for i,e in enumerate(Nbins):
    sub[i].hist(titanic["Age"], bins = e)
    sub[i].set_xlabel("Age")
```

#### Whether the bin size affects your plot? Why?

The bin size does not affect the plot size but it affects the axes. In this example here, the more bins used, the smaller the groups are, so the y axes gets smaller values.

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the appropriate plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.

*Hint: One of the ways involves using Seaborn*

```python
#your code-1st way: Plotting linear line regression with plt
x = titanic["Age"]
y = titanic["Fare"]
plt.plot(x, y, "m*") #sacar una línea de regresión??
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, Linewidth=4, color="g")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Relationship between Age and Fare");
```

```python
# your code-2nd way
sns.lmplot(x="Age", y="Fare", data=titanic);
```

```python
sns.jointplot(x='Fare',y='Age', kind="reg",data=titanic);
```
